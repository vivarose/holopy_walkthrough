```python
import holopy as hp
from holopy.core import Optics
from holopy.fitting import Model, par, fit
#from holopy.scattering.scatterer.sphere import LayeredSphere
from holopy.scattering.theory import Mie
#from importyaml import importyaml
import matplotlib.pyplot as plt     # for plt.title
import os
import datetime

myhome = '/home/viva'

## I am going to fit the hologram movie of a Janus swimmer
## The data is located at
folder = os.path.join(myhome, 'group/vesicles/2014-02-27/janus_in_peroxide/44/')

## First it's good to watch a movie of these raw hologram images at 
## /group/manoharan2/vesicles/2014-02-27/janus_in_peroxide/movies/44_holo_defocused.avi,
## But here we can just look at a representative image

# Load the image, including information about the scalebar, 
# the laser wavelength, and the index of refraction of the hydrogen peroxide in water
img0000_orig = hp.load(folder + 'image0000.tif',
                       spacing = 0.1,
                       optics = Optics(wavelen=.65845,
                                       index = 1.3518,
                                       polarization=(1,0)))
```

```python
# Tell python to display images in this notebook
%pylab inline
```

```python
# Display the image
hp.show(img0000_orig)
plt.title('image0000.tif')
```

```python
# There's a lot of stuff in the background that isn't moving. 
# We only want to focus on the one swimming particle. 
# A good way to remove the background is to take an average of all the images in the folder.

bg_filename = os.path.join(myhome, 'group/vesicles/2014-02-27/janus_in_peroxide/44_averaged.yaml')

if os.path.exists(bg_filename):
    # Load background image
    bg = hp.load(bg_filename)
else:
    # Calculate background image (might be slow) and save it for next time
    bg = hp.core.io.average_images(images=folder)   # This function averages images
                # it may be better to take a median than a mean
    hp.save(bg_filename, bg)
    
hp.show(bg)
plt.title('Background image')
```

```python
# Then we can see how the image looks when we remove the parts of the image 
# that don't move over the course of the movie

img0000 = img0000_orig/bg         # Background division
hp.show(img0000)
plt.title('img0000, background divided')
```

```python
# The moving particle is much clearer now, 
# but there's still some distracting stuff. 
# We can focus in on the best parts of the image by cropping.
# Note to self: x and y are swapped from their usual axes.

img0000cropped = hp.core.process.normalize(
    hp.core.marray.subimage(img0000,     # image
                            [400,500],   # center of image [in pixels]
                            350))        # size of cropped image [in pixels]

hp.show(img0000cropped)
plt.title('img0000, cropped')
```

```python
img0000cropped.min()  
# Make sure the image is positive before running the propagate command. 
# I had a problem when I used to subtract the background instead of dividing..
```

```python
# One of the traditional ways that holograms have been used is reconstruction. 
# This is analogous to propagating light through a film to recreate a 3D image of an object.
# With digital holography, we can choose how far to propogate the light.
# This means that we can essentially change the focal plane of the image.

p = hp.propagate(img0000cropped, 20)
```

```python
hp.show(p)
plt.title('Reconstructed from data,\n' + u"\u0394" + 'z = 20 microns')

# It's not perfect! Tom doesn't like reconstruction. Vinny does.
# I changed the depth until it looked like I had focused on the Janus particle.
# It's not quite the same image as we would get if we just changed focus experimentally,
# but it's helpful for guessing the distance between the focal plane and the object.
```

```python
# In order to fit to this data, we need to simulate a hologram, 
# then compare the simulation to the data and adjust the simulation iteratively 
# until we have a good match. The fitting algorithm does most of the work, 
# but we need to start with a good guess.

from holopy.scattering.scatterer import Sphere, Spheres
from holopy.scattering.theory import Mie, Multisphere
from holopy.fitting import fit, par, Model
from holopy.core import ImageSchema, Optics

# We can simulate the hologram for a sphere in a 3d position with an index of refraction and a radius.
simulated_sphere = Sphere(center = [38,52,20], 
                          n = 1.5917,   # refractive index of polystyrene
                          r=0.5)
simulated_sphere_hologram = Mie.calc_holo(simulated_sphere, scaling=0.6, schema = img0000cropped)
# An image schema has information about the size of the image, 
# number of pixels, and the laser. 
# We don't need to input the schema in manually if we tell holopy it's the same as img0000cropped.
# "scaling" is also called "alpha", and Aaron has been investigating it.

hp.show(simulated_sphere_hologram)
plt.title('My guess (simulation)')

hp.show(img0000cropped)
plt.title('img0000, cropped (data)')

# I didn't come up with a good guess right from the beginning, 
# I adjusted the numbers I input for the center of the particle until it looked fairly good.
```

```python
# To find the x,y,z coordinates of the sphere, 
# we parameterize the center and radius of the sphere 
# and use a fitting algorithm.

parameterized_sphere = Sphere(
                        center = [par(guess = 38, limit = [10,70]),
                                  par(guess = 52, limit = [20,80]),
                                  par(guess = 20, limit = [.1,100])],                                   
                        n = 1.5917,   # refractive index of polystyrene
                        r= par(guess = 0.5, limit = [.1, 1]))

model = Model(parameterized_sphere, 
              Mie.calc_holo,
              alpha = par(.6, [.3, 1])) 

result = fit(model, img0000cropped)

```

```python
guess = Mie.calc_holo(model.scatterer.guess, scaling = 0.6, schema = img0000cropped)
res = result.fitted_holo(schema = img0000cropped)

hp.show(res - guess)
plt.title('Fit result minus guess')
hp.show(img0000cropped - guess)
plt.title('Data minus guess')
hp.show(img0000cropped - res)
plt.title('Data minus fit result')
```

```python
# If it's not yet CLEAR that the fit result shown above is much better than the guess was, 
# look at the scale on the right, not just the appearance of the subtracted images.

# It looks properly centered, but there's something a little off about the rings.
# Last time I didn't allow the index to vary. 
# Maybe I should have, especially considering that 
# there is platinum on the polystyrene particle that I'm ignoring.
# Sometimes using the fit results and fitting again with different parameters can help.
# I'll try fitting the depth, index, and radius to see if I can get a better result.


parameterized_sphere2 = Sphere(
                        center = [result.scatterer.center[0],
                                  result.scatterer.center[1],
                                  par(guess = result.scatterer.center[2], limit = [15,25])],                                   
                        n = par(guess = 1.5917, limit = [1.5,1.7]),   
                        r = par(guess = result.scatterer.r, limit = [.1, 1]))

model2 = Model(parameterized_sphere2, 
              Mie.calc_holo,
              alpha = par(result.alpha, [.1, 1])) 

start_time = datetime.datetime.now()
result2 = fit(model2, img0000cropped)  # do the fit
end_time = datetime.datetime.now()

how_long = end_time - start_time
how_long.seconds      # display the time it took to do the fit
```
